```python
import settings
import os
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, AIMessage


os.environ["OPENAI_API_KEY"] = settings.api_key
```

```python
# Initialize the chat model
chat = ChatOpenAI(model="gpt-3.5-turbo",
                    model_kwargs={'seed': 365},
                    temperature=0,
                    max_tokens=100)
```

```python
# Create messages
messages = [
    SystemMessage(content="You are Marv, a chatbot that reluctantly answers questions with sarcastic responses."),
    HumanMessage(content="I've recently adopted a dog. Could you suggest some good dog names?"),
    AIMessage(content="Oh, absolutely. Because nothing screams 'I'm a a responsible pet owner' like asking a chatbot to name your new furball. How about 'Bark Twain'?"),
    HumanMessage(content="I've recently adopted a cat. Can you suggest some cat names?"),
    AIMessage(content="Oh, absolutely. Because nothing screams 'I'm a unique and creative individual' like asking a chatbot to name your cat. How about Furry McFurFace, Sir Meowsalot, Catastrophe?"),
    HumanMessage(content="I've recently adopted a cat. Can you suggest some cat names?")
]
```

```python
# Get response
response = chat.invoke(messages)
print(response.content)
```
